```python
import torchaudio
import whisper
import tempfile
import torchaudio
```

```python
# carrega dataset (faz download se necessário)
dataset = torchaudio.datasets.LIBRISPEECH(root="~/.cache", url="test-clean", download=True)
```

```python
# escolher índice de teste
idx = 0
waveform, sample_rate, reference_text, *_ = dataset[idx]  # waveform: Tensor [channels, frames]
```

```python
# salvar WAV temporário
tmp = tempfile.NamedTemporaryFile(suffix=".wav", delete=False)
tmp_path = tmp.name
tmp.close()
torchaudio.save(tmp_path, waveform, sample_rate)
```

```python
# transcrever com whisper
model = whisper.load_model("base")
res = model.transcribe(tmp_path)
```

```python
print("Referencia:", reference_text)
print("Predicao :", res["text"])
```
