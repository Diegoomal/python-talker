Create audio

```python
import torch
from transformers import BarkModel, AutoProcessor

device = "cuda"
```

```python
model = BarkModel.from_pretrained("suno/bark-small").to(device)
processor = AutoProcessor.from_pretrained("suno/bark")
```

```python
text_prompt = "Yes, add it."
```

```python
inputs = processor(text_prompt, return_tensors="pt").to(device)

with torch.no_grad():
    speech_output = model.generate(**inputs)
```

Save Audio

```python
filename = "3.wav"
```

```python
import scipy.io.wavfile as wav # type: ignore

audio = speech_output.cpu().numpy().squeeze()
wav.write(filename, model.generation_config.sample_rate, audio)
```

Convert to mp3

```python
from pydub import AudioSegment # type: ignore

AudioSegment.from_wav(filename).export(filename, format="mp3")
```
