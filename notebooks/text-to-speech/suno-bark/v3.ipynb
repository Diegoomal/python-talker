```python
import torch                                        # type: ignore
from transformers import BarkModel, AutoProcessor   # type: ignore
import scipy.io.wavfile as wav                      # type: ignore

device = "cuda"

model = BarkModel.from_pretrained("suno/bark").to(device)
processor = AutoProcessor.from_pretrained("suno/bark")

text_prompt = [
    "Hi, I want to place a pizza order.",
    "I want one large pizza.",
    "Yes.",
    "Yes, add it.",
    "No, that's all."
]
```

```python
for idx, prompt in enumerate(text_prompt):

    print(f"Generating audio for prompt {idx}: {prompt}")

    inputs = processor(prompt, voice_preset="en_speaker_6", return_tensors="pt").to(device)

    with torch.no_grad():
        audio = model.generate(**inputs)

    audio = audio.cpu().numpy().squeeze()
    wav.write(f"{idx}.wav", model.generation_config.sample_rate, audio)
```
